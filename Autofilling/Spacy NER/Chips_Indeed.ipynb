```python
import time
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
#from seleniumbase import SB

indeed_urls = ['https://pk.indeed.com/jobs?q=Data+Science&l=Karachi&pp=gQAAAAAAAAAAAAAAAAACEWW-2QADAAABAAA&vjk=106e278642a62472']
```

```python
from bs4 import BeautifulSoup
import time
import pandas as pd

driver = webdriver.Chrome()
driver.get(indeed_urls[0])

total_links = []

count = 0
while True:
    res = driver.page_source.encode('utf-8')
    soup = BeautifulSoup(res, "html.parser")
    
    count += 1
    print(count)
    
    job_boxes = soup.find_all("li", class_="css-5lfssm eu4oa1w0")
    total_links.extend(job_boxes)
    pagination_next = soup.find("a", {"data-testid": "pagination-page-next"})
    if pagination_next:
        new_url = "https://pk.indeed.com" + pagination_next["href"]
        driver.get(new_url)
        time.sleep(10)
    else:
        break

print(total_links)
print(len(total_links))
```

```python
Data_frame = pd.DataFrame({'Title':[],"Company":[],"Description":[], "Salary" : [], "Deadline":[],"Locations":[], "Mode" : [], "Type" : [],
                          "Gender" : [], "Experience" : [], "Field_Of_Interests" : [], "Source" : []})
            #Adding Source to know where we got this information from
count = 0
for link in total_links:
    #print(link)
    #print("------------------------------------------------------------------------------------------")
    Data_record = []
    element = link.find("a", class_="jcs-JobTitle css-jspxzf eu4oa1w0", href=True)
    #print(element)
#     if count == 3:
#         break;
    if element:
        job_url = "https://pk.indeed.com" + element["href"]
        print(job_url)
        
        driver.get(job_url)
        time.sleep(5)
        
        res = driver.page_source.encode('utf-8')
        soup = BeautifulSoup(res, "html.parser")
        #print(soup)
        
        #Title here
        title = soup.find('h1', class_='jobsearch-JobInfoHeader-title css-1hwk56k e1tiznh50')
        if title:
            #print(title.find('span').text)
            Data_record.append(title.find('span').text)
        else:
            Data_record.append(None)
        
        #Company here
        company = soup.find('a', class_='css-1f8zkg3 e19afand0')
        if company:
            #print(company.text)
            Data_record.append(company.text)
        else:
            Data_record.append(None)
        
        #Description here
        clean_description = ""
        descriptiondiv = soup.find('div', {'id': 'jobDescriptionText'}) #Can improve description by adding strings of Ability to commute and Ability to relocate
        if descriptiondiv:
            description = descriptiondiv.find_all('p')
            for des in description:
                clean_description += (des.text + " ")
            #print(clean_description)
            Data_record.append(clean_description)
        else:
            Data_record.append(None)
                
        #Salary here
        salary = soup.find('div', class_='css-tvvxwd ecydgvn1') #Salary and Type are getting interchanged (FIX THIS)
        if salary:
            #print(salary.text)
            Data_record.append(salary.text)
        else:
            Data_record.append(None)
        
        #Deadline here
        Data_record.append(None)
        
        #Location here
        location = soup.find('div', class_='css-6z8o9s eu4oa1w0')
        if location:
            #print(location.text)
            Data_record.append(location.text)
        else:
            Data_record.append(None)
            
        #Mode here
        Data_record.append(None)
        
        #Type here
        type_ = soup.find_all('div', class_='css-tvvxwd ecydgvn1')
        if len(type_) > 1:
            #print(type_[1].text)
            Data_record.append(type_[1].text)
        else:
            Data_record.append(None)
        
        #Gender here
        Data_record.append(None)
        
        #Experience here
        Data_record.append(None)
        
        #Field Of Interests here
        Data_record.append(None)
        
        #Source here
        Data_record.append(job_url)
        
        
#         count += 1
        #print(count)
        print(Data_record)
        print(len(Data_record))
        Data_frame = Data_frame.append(pd.Series(Data_record, index=Data_frame.columns), ignore_index=True)
    else:
        continue
```

```python
csv_filename = 'Chips_Indeed_Data.csv'

Data_frame.to_csv(csv_filename, index=False)
```
